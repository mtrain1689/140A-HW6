### import library

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

# PART A

```python
data = np.loadtxt("data/hw6_file.txt")
data -= np.mean(np.mean(data))

fig = plt.imshow(data, extent = [0, 12e-9, 0, 12e-9])
plt.xlabel("Distance along x axis")
plt.ylabel("Distance along y axis")

plt.vlines(ymin = 0, ymax = 1.2e-8, x = 0.5e-8, color = "red")
plt.vlines(ymin = 0, ymax = 1.2e-8, x = 0.54e-8, color = "red")
plt.hlines(xmin = 0, xmax = 1.2e-8, y = 0.05e-8, color = "purple")
plt.hlines(xmin = 0, xmax = 1.2e-8, y = 0.08e-8, color = "purple")


txt = "Scanning tunnel microscopy data of unknown material with reciprocal properties"
plt.figtext(0.5, -0.05, txt, wrap = True, horizontalalignment = "center")
```

From the above image, the smaller atomic lattice constant (a) seems to be approximately 0.5 nm. Length of superlattice is 3a. Both are rhombohedral.
